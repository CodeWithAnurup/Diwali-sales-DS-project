# DATA ANALYSIS

```python
# Importing the Pandas library for data manipulation and analysis
import pandas as pd

# Importing the NumPy library for numerical computations and handling arrays
import numpy as np

# Importing the Matplotlib library for data visualization
import matplotlib.pyplot as plt

# Importing the Seaborn library for advanced statistical data visualization
import seaborn as sns
```

```python
# Loading the CSV file into a DataFrame
df = pd.read_csv(r"C:\Users\anuru\Desktop\Dewali sales\Diwali Sales Data.csv", encoding='latin1')
```

```python
# Displaying the DataFrame 'df'
df
```

```python
# Displaying the dimensions of the DataFrame
# df.shape returns a tuple (number of rows, number of columns) in the DataFrame.
df.shape
```

```python
# Displaying the first 5 rows of the DataFrame
df.head()
```

```python
# Display a concise summary of the DataFrame
df.info()
```

```python
# Dropping the 'Status' and 'unnamed1' columns from the DataFrame
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
# Checking for missing values in each column of the DataFrame
pd.isnull(df).sum()
```

```python
# Removing rows with missing values from the DataFrame
df.dropna(inplace=True)
```

```python
# Checking the number of missing values in each column of the DataFrame
pd.isnull(df).sum()
```

```python
# Converting the 'Amount' column to integer data type
df['Amount']=df['Amount'].astype('int')
```

```python
# Checking the data type of the 'Amount' column
df['Amount'].dtypes
```

```python
# Displaying the list of column names in the DataFrame
df.columns
```

```python
# Generating descriptive statistics
df[['Age','Orders','Amount']].describe()
```

#   GENDERWISE DATA ANALYSIS

```python
#PLOTING A BAR CHART FOR GENDER AND ITS COUNT
ax = sns.countplot(x='Gender', data=df)

# Add count labels on top of bars
for bars in ax.containers:
    ax.bar_label(bars)  # <- Corrected this line

plt.show()
```

```python
#PLOTING A BAR CHAT FOR GENDER VS TOTAL AMOUNT

sales_gender=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gender)
```

```python
df.head()
```

# AGEWISE DATA ANALYSIS

```python
#PLOTING A BARCHAT FOR AGE AND ITS COUNT 

ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#PLOTING A BARCHAT FOR AGE GROUP VS TOTAL AMOUNT

sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

# STATEWISE DATA ANALYSIS

```python
#PLOTING A BARCHAT FOR STATEWISE ORDERS

sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders').head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
#PLOTING A BARCHAT FOR STATE VS TOTAL AMOUNT

sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount').head(10)
sns.barplot(x='State',y='Amount',data=sales_state)
```

# MARITAL STATUS DATA ANALYSIS

```python
#PLOTING A BARCHAT FOR MARITAL STATUS AND ITS COUNT

ax=sns.countplot(data=df,x='Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#PLOTING A BARCHAT FOR MARITIAL STATUS VS TOTAL AMOUNT

sales_marital=df.groupby(['Marital_Status'],as_index=False)['Amount'].sum().sort_values(by='Amount').head(10)
sns.barplot(x='Marital_Status',y='Amount',data=sales_marital)
```

# OCCUPATIONWISE DATA ANALYSIS

```python
#PLOTING A BARCHAT FOR OCCUPATION AND ITS COUNT

ax=sns.countplot(data=df,x='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#PLOTING A BARCHAT FOR OCCUPATION VS TOTAL AMOUNT

sales_occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum()
sns.barplot(x='Occupation',y='Amount',data=sales_occupation)
```

# PRODUCT CATEGORYWISE DATA ANALYSIS

```python
#PLOTING A BARCHAT FOR PRODUCT CATEGORY AND ITS COUNT

ax=sns.countplot(data=df,x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#PLOTING A BARCHAT FOR PRODUCT CATEGORY VS TOTAL AMOUNT

sales_product=df.groupby(['Product_Category'],as_index=False)['Amount'].sum()
sns.barplot(x='Product_Category',y='Amount',data=sales_product)
```
